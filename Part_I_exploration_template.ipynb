## Introduction


The Programme for International Student Assessment ('PISA') is a global study conducted in member and non-member countries by the Organization for Economic Cooperation and Development to evaluate educational systems by measuring 15-year-old school pupils' scholastic performance in mathematics, science, and reading. and this dataset is the outcome of the 2012 study





## Preliminary Wrangling


```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
base_color = sb.color_palette()[0]
%matplotlib inline
```

```python
pisa = pd.read_csv('pisa2012.csv', encoding = "ISO-8859-1")
```

```python
# rename the dataframe columns
pisa = pisa.rename(columns={'CNT': 'Country', 
                            'PV1MATH': 'Math_score', 
                            'PV1READ': 'Reading_score', 
                            'PV1SCIE': 'Science_score', 
                            'ST04Q01': 'Gender', 
                            'ST01Q01': 'Grade', 
                            'SCHOOLID': 'School_ID',
                            'STIDSTD': 'Student_ID',
                            'ESCS' : 'Socio_economic_status',
                            'ST29Q04': 'Students_enjoy_math',
                            'ST57Q01':'Homework_hours',
                            'INTMAT':'Math_interest',
                            'LMINS':'Language_in_school_min',
                            'MMINS':'Math_in_school_min',
                            'SMINS':'Science_in_school_min',
                            'SUBNORM':'Math_motivation',
                            'ST35Q04': 'Parents_math_important',
                            'ST35Q05':'Parents_math_for_career',
                            'ST13Q01': 'Mother Education', 'ST17Q01': 'Father Education'})
# keeping only the relevant fields

pisa = pisa[['Country', 
                            'Math_score', 
                            'Reading_score', 
                            'Science_score', 
                            'Gender', 
                            'Grade', 
                            'School_ID',
                            'Student_ID',
                            'Socio_economic_status',
                            'Students_enjoy_math',
                            'Homework_hours',
                            'Math_interest',
                            'Language_in_school_min',
                            'Math_in_school_min',
                            'Science_in_school_min',
                            'Math_motivation',
                            'Parents_math_important',
                            'Parents_math_for_career',
                            'Mother Education', 
                            'Father Education',
                            'WEALTH']]
#re encoding Father Education
pisa['Father Education'] = pisa['Father Education'].replace({'<ISCED level 3A> ':'3A', 
                                             '<ISCED level 3B, 3C> ':'3B', 
                                             '<ISCED level 2> ':'secondary',
                                             '<ISCED level 1> ':'primary',
                                             'He did not complete <ISCED level 1> ':'pre-primary'}) 
#re encoding Mother Education
pisa['Mother Education'] = pisa['Mother Education'].replace({'<ISCED level 3A> ':'3A', 
                                             '<ISCED level 3B, 3C> ':'3B', 
                                             '<ISCED level 2> ':'secondary',
                                             '<ISCED level 1> ':'primary',
                                             'She did not complete <ISCED level 1> ':'pre-primary'})


#removing nulls
pisa.dropna(inplace=True)

#removing duplicates
pisa.drop_duplicates(inplace=True)

#checking
pisa.info()

# saving it into a new csv file
pisa.to_csv('pisa2012clean.csv')
```

## Univariate Exploration



> ### Which countries have the most students, on average?

```python
order = pisa['Country'].value_counts().index 
max = pisa['Country'].value_counts()[0]
b = np.arange(0, max, 500)
plt.figure(figsize=[10,20])
sb.countplot(data=pisa, y = 'Country', color = base_color, order=order);
plt.xticks(b);
plt.title('Students Distributions across countries', size=20);
```

> Mexico has the most students participants across the country.

> ### what is the Distribution of different subjects scores

```python
pisa[['Math_score', 'Reading_score', 'Science_score',]].hist(figsize=[20,20]);
```

> It appears that the distribution of science scores is similar to that of math scores.

> ### how many students are male or female

```python
index = pisa['Gender'].value_counts()
plt.pie(index.values,  labels= index.index, autopct='%1.0f%%', startangle=90 )
plt.axis('equal')  
plt.title('Students\' Gender Count')
plt.show()
```

> It appears that females outnumber males in participation.

> ### what is the overall Grade Distribution

```python
# print the grade distro
plt.figure(figsize=[10,6])
sb.distplot(pisa['Grade'])
plt.xlabel('Grade Distribution')
plt.show()

pisa['Grade'].value_counts()
```

> it seems that there is an outlier

```python
#dropping all the Grades higher than 14.
pisa = pisa[pisa['Grade'] <= 14]
pisa['Grade'].value_counts()
```

> ### What is each country's average grade?

```python
grade_mean = (pisa.groupby('Country')['Grade'].mean()).sort_values(ascending=False)
plt.figure(figsize=[16,20])
sb.barplot( y= grade_mean.index, x = grade_mean.values, color ='#f26e1d')
plt.title('Average Grade by each Country', size=35)
plt.xlabel('Grades');
```

> Among the other countries, the United Kingdom and New Zealand have the highest average score.

> ### What is the distribution of Socio_economic_status?

```python
bins = np.arange(pisa['Socio_economic_status'].min() - 1, pisa['Socio_economic_status'].max()+1, 0.1)
plt.figure(figsize=[20, 16])
plt.hist(data = pisa, x = 'Socio_economic_status', bins = bins)
plt.xlabel('socioeconomic')
plt.show()
```

> It appears that the majority of countries have a negative socioeconomic index score

> ### how much time is spent at home doing homework

```python
bin_edges = np.arange(-0.48, pisa.Homework_hours.max()+1, 1)
plt.figure(figsize=[8, 5])
plt.hist(data = pisa, x = 'Homework_hours', bins = bin_edges, rwidth = 0.8)
plt.xticks(np.arange(0, pisa.Homework_hours.max()+1, 1));
```

We can infer from the graph above that most students do homwork for 1-3 hours.

## Bivariate Exploration



> ### Wealth and socioeconomic statusdo they correlate in any way

```python
plt.figure (figsize = [20,17])
sb.regplot(x= 'WEALTH', y='Socio_economic_status', data = pisa, color =base_color, scatter_kws={'alpha':1/40} );
pisa['WEALTH'].corr(pisa['Socio_economic_status'])
```

> There is a fairly Strong positive correlation of 0.685261572473708 between the two.

> ### How much of an impact can wealth have on grades?

```python
plt.figure (figsize = [20,12])
sb.regplot(y= 'WEALTH', x='Grade', data = pisa, color =base_color, x_jitter=0.3, scatter_kws={'alpha':1/40});
pisa['WEALTH'].corr(pisa['Grade'])
```

> there is a negligible positive correlation between the two, which effectively negates any meaningful relationship between wealth and grade.

>  ### How much of an impact can Socio_economic_status have on grades?

```python
plt.figure (figsize = [20,12])
sb.regplot(y= 'Socio_economic_status', x='Grade', data = pisa, color =base_color, x_jitter=0.3, scatter_kws={'alpha':1/40});
pisa['Socio_economic_status'].corr(pisa['Grade'])
```

Given that wealth and socioeconomic status have a strong positive correlation, it makes sense that their correlation with academic performance is minimal.
